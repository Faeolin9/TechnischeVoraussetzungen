# Lösungen zur Abschlussübung: Analyse des kalifornischen Wohnungsdatensatzes

## Setup: Bibliotheken importieren und Daten laden

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Load the dataset
housing = fetch_california_housing(as_frame=True)
housing_df = housing.frame
```

### Aufgabe 1: Was ist der durchschnittliche (mittlere) `MedHouseVal` (Medianer Hauswert) für alle Bezirke?

```python
average_house_value = housing_df['MedHouseVal'].mean()
print(f"Der durchschnittliche mediane Hauswert beträgt: {average_house_value:.2f} (in 100.000 $)")
```

### Aufgabe 2: Filtern Sie den DataFrame, um nur die Bezirke anzuzeigen, in denen das `HouseAge` größer als 50 Jahre ist. Wie viele solcher Bezirke gibt es?

```python
old_houses_df = housing_df[housing_df['HouseAge'] > 50]
print(f"Es gibt {len(old_houses_df)} Bezirke mit einem medianen Hausalter von mehr als 50 Jahren.")
old_houses_df.head()
```

### Aufgabe 3: Erstellen Sie eine neue Spalte namens `RoomsPerPerson`, indem Sie die Spalte `AveRooms` durch die Spalte `AveOccup` dividieren.

```python
housing_df['RoomsPerPerson'] = housing_df['AveRooms'] / housing_df['AveOccup']
housing_df[['AveRooms', 'AveOccup', 'RoomsPerPerson']].head()
```

### Aufgabe 4 (Herausforderung): Erstellen Sie ein Streudiagramm, um die Beziehung zwischen `MedInc` (Medianes Einkommen) auf der x-Achse und `MedHouseVal` (Medianer Hauswert) auf der y-Achse zu visualisieren. Geben Sie dem Diagramm einen Titel und Beschriftungen.

```python
plt.figure(figsize=(10, 6))
plt.scatter(housing_df['MedInc'], housing_df['MedHouseVal'], alpha=0.5)
plt.title('Medianes Einkommen vs. Medianer Hauswert in Kalifornien')
plt.xlabel('Medianes Einkommen (in 10.000 $)')
plt.ylabel('Medianer Hauswert (in 100.000 $)')
plt.grid(True)
plt.show()
```
